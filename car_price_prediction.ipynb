##### Fixing invalid values
-  ‘∏’∂’Ø’•÷Ä’∏÷Ç’©’µ’°’∂ ’°’∂’∏÷Ç’∂’® ’Ω’µ’∏÷Ç’∂’°’Ø’∏÷Ç’¥ ’∏÷Ç’≤’≤’°’£÷Ä’°’Ø’°’∂ ’Ω’≠’°’¨ ’Ø’°:

    - `Nissan` = `nissan`

    - `toyota` = `toyouta`
    - `vokswagen` = `volkswagen` =  `vw`

```python
cars.CompanyName = cars.CompanyName.str.lower()

def replace_name(a,b):
    cars.CompanyName.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

cars.CompanyName.unique()
```

### Step 4 : Deriving new features(’Ü’∏÷Ä ’∞’°’ø’Ø’°’∂’´’∑’∂’•÷Ä’´ ’Ω’ø’°÷Å’∏÷Ç’¥)

```python
#Fuel economy
cars['fueleconomy'] = (0.55 * cars['citympg']) + (0.45 * cars['highwaympg'])
```

Highway mpg-’® ’•÷Ä’•÷Ñ ’æ’°÷Ä’Ø’°’∂’´’∑’∂’•÷Ä’´÷Å ’¥’•’Ø’∂ ’ß, ’∏÷Ä’® ’Ñ’´’°÷Å’µ’°’¨ ’Ü’°’∞’°’∂’£’∂’•÷Ä’´ ’á÷Ä’ª’°’Ø’° ’¥’´’ª’°’æ’°’µ÷Ä’´ ’∫’°’∑’ø’∫’°’∂’∏÷Ç’©’µ’°’∂ ’£’∏÷Ä’Æ’°’Ø’°’¨’∏÷Ç’©’µ’∏÷Ç’∂’® (EPA) ÷Ö’£’ø’°’£’∏÷Ä’Æ’∏÷Ç’¥ ’ß ’¥’•÷Ñ’•’∂’°’µ’´ ’æ’°’º’•’¨’´÷Ñ’´ ’≠’∂’°’µ’∏’≤’∏÷Ç’©’µ’∏÷Ç’∂’® ’∏÷Ä’∏’∑’•’¨’∏÷Ç ’∞’°’¥’°÷Ä:

City MPG: the score a car will get on average in city conditions, with stopping and starting at lower speeds.

```python
#Binning the Car Companies based on avg prices of each Company.
cars['price'] = cars['price'].astype('int')
temp = cars.copy()
table = temp.groupby(['CompanyName'])['price'].mean()
temp = temp.merge(table.reset_index(), how='left',on='CompanyName')
bins = [0,10000,20000,40000]
cars_bin=['Budget','Medium','Highend']
cars['carsrange'] = pd.cut(temp['price_y'],bins,right=False,labels=cars_bin)
cars.head()
```

### Step 6 : Dummy Variables

‘ø’•’≤’Æ ÷É’∏÷É’∏’≠’°’Ø’°’∂’® ’•÷Ä’Ø’∏÷Ç’°’Ø’°’∂ ÷É’∏÷É’∏’≠’°’Ø’°’∂ ’ß, ’∏÷Ä’∂ ’®’∂’§’∏÷Ç’∂’∏÷Ç’¥ ’ß 0 ’Ø’°’¥ 1 ’°÷Ä’™’•÷Ñ’ù ÷Å’∏÷Ç’µ÷Å ’ø’°’¨’´’Ω  ’Ø’°’ø’•’£’∏÷Ä’´’Ø(categorical) ’ß÷Ü’•’Ø’ø’´ ’¢’°÷Å’°’Ø’°’µ’∏÷Ç’©’µ’∏÷Ç’∂’® ’Ø’°’¥ ’°’º’Ø’°’µ’∏÷Ç’©’µ’∏÷Ç’∂’®:

```python
# Defining the map function
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
# Applying the function to the cars_lr

cars_lr = dummies('fueltype',cars_lr)
cars_lr = dummies('aspiration',cars_lr)
cars_lr = dummies('carbody',cars_lr)
cars_lr = dummies('drivewheel',cars_lr)
cars_lr = dummies('enginetype',cars_lr)
cars_lr = dummies('cylindernumber',cars_lr)
cars_lr = dummies('carsrange',cars_lr)
```

RFE-’∂ ’∂’∑’°’∂’°’Ø’∏÷Ç’¥ ’ß Recursive Feature Elimination: ‘¥’° ’°’º’°’∂’±’∂’°’∞’°’ø’Ø’∏÷Ç’©’µ’∏÷Ç’∂’∂’•÷Ä’´ ’®’∂’ø÷Ä’∏÷Ç’©’µ’°’∂ ’°’¨’£’∏÷Ä’´’©’¥ ’ß, ’∏÷Ä’∂ ÷Ö’£’ø’°’£’∏÷Ä’Æ’æ’∏÷Ç’¥ ’ß ’ø’æ’µ’°’¨’∂’•÷Ä’´ ’¢’°’¶’°’µ’´÷Å ’∞’°’ø’Ø’°’∂’´’∑’∂’•÷Ä’´ ’•’∂’©’°’¢’°’¶’¥’∏÷Ç’©’µ’∏÷Ç’∂ ’®’∂’ø÷Ä’•’¨’∏÷Ç ’∞’°’¥’°÷Ä: RFE-’∂ ’°’∑’≠’°’ø’∏÷Ç’¥ ’ß’ù ’°’∂’®’∂’§’¥’•’ª ’∞’•’º’°÷Å’∂’•’¨’∏’æ ’°’¥’•’∂’°÷Ñ’´’π ’Ø’°÷Ä÷á’∏÷Ä ’∞’°’ø’Ø’°’∂’´’∑’∂’•÷Ä’® ’ø’æ’µ’°’¨’∂’•÷Ä’´ ’∑’ø’•’¥’°÷Ä’°’∂’´÷Å ÷á ’æ’•÷Ä’°’∫’°’ø÷Ä’°’Ω’ø’•’¨’∏’æ ’¥’∏’§’•’¨’®: ‘≥’∏÷Ä’Æ’®’∂’©’°÷Å’® ’Ø÷Ä’Ø’∂’æ’∏÷Ç’¥ ’ß ’°’µ’∂÷Ñ’°’∂ ’™’°’¥’°’∂’°’Ø, ’¥’´’∂’π÷á ’∞’°’Ω’∂’•’∂ ’£’∏÷Ä’Æ’°’º’∏÷Ç’µ’©’∂’•÷Ä’´ ÷Å’°’∂’Ø’°’¨’´ ÷Ñ’°’∂’°’Ø’∏÷Ç’©’µ’∏÷Ç’∂’®:

```python
#RFE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm       #Statsmodels API-’∂ ’∞’¶’∏÷Ä ’£÷Ä’°’§’°÷Ä’°’∂ ’ß ’æ’´’≥’°’Ø’°’£÷Ä’°’Ø’°’∂ ’¥’∏’§’•’¨’°’æ’∏÷Ä’¥’°’∂ ÷á ’ø’æ’µ’°’¨’∂’•÷Ä’´ ’æ’•÷Ä’¨’∏÷Ç’Æ’∏÷Ç’©’µ’°’∂ ’∞’°’¥’°÷Ä:
from statsmodels.stats.outliers_influence import variance_inflation_factor
#Variance_inflation_factor() ÷Ü’∏÷Ç’∂’Ø÷Å’´’°’∂ statsmodels.stats.outliers_influence ’£÷Ä’°’§’°÷Ä’°’∂’´÷Å ÷Ö’£’ø’°’£’∏÷Ä’Æ’æ’∏÷Ç’¥ ’ß ’¥’´ ’∑’°÷Ä÷Ñ ’°’∂’Ø’°’≠ ÷É’∏÷É’∏’≠’°’Ø’°’∂’∂’•÷Ä’´ ’∞’°’¥’°÷Ä ’§’´’Ω’∫’•÷Ä’Ω’´’°’µ’´ ’£’∂’°’≥’´ ’£’∏÷Ä’Æ’°’Ø’´÷Å’® (VIF) ’∞’°’∑’æ’°÷Ä’Ø’•’¨’∏÷Ç ’∞’°’¥’°÷Ä:
```

### Building model using statsmodel, for the detailed statistics(‘ø’°’º’∏÷Ç÷Å’¥’°’∂ ’¥’∏’§’•’¨’ù ÷Ö’£’ø’°’£’∏÷Ä’Æ’•’¨’∏’æ statsmodel, ’¥’°’∂÷Ä’°’¥’°’Ω’∂ ’æ’´’≥’°’Ø’°’£÷Ä’∏÷Ç’©’µ’°’∂ ’∞’°’¥’°÷Ä)

```python
X_train_rfe = X_train[X_train.columns[rfe.support_]]
X_train_rfe.head()
```

```python
def build_model(X,y):
    X = sm.add_constant(X) #Adding the constant(‘±’æ’•’¨’°÷Å’∂’•’¨’∏’æ ’∞’°’Ω’ø’°’ø’∏÷Ç’∂)
    lm = sm.OLS(y,X).fit() # fitting the model( ’¥’∏’§’•’¨’´ ’∞’°’¥’°’∫’°’ø’°’Ω’≠’°’∂’°÷Å’∏÷Ç’¥)
    print(lm.summary())   # model summary(’¥’∏’§’•’¨’´ ’°’¥÷É’∏÷É’∏÷Ç’¥)
    return X
    
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
#Scaling the test set(’ì’∏÷Ä’±’°÷Ä’Ø’¥’°’∂ ’∞’°’æ’°÷Ñ’°’Æ’∏÷Ç’´ ’¥’°’Ω’∑’ø’°’¢’°’æ’∏÷Ä’∏÷Ç’¥)
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'horsepower','fueleconomy','carlength','carwidth','price']
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])
#’°’µ’Ω ’Ø’∏’§’® ÷É’∏’≠’°’Ø’•÷Ä’∫’∏÷Ç’¥ ’ß ’ø’æ’µ’°’¨’∂’•÷Ä’® df_test DataFrame-’∏÷Ç’¥’ù ÷Ö’£’ø’°’£’∏÷Ä’Æ’•’¨’∏’æ scaler ÷Ö’¢’µ’•’Ø’ø’®: Scaler ÷Ö’¢’µ’•’Ø’ø’® sklearn.preprocessing.StandardScaler ÷Ö’¢’µ’•’Ø’ø ’ß, ’∏÷Ä’∂ ÷Ö’£’ø’°’£’∏÷Ä’Æ’æ’∏÷Ç’¥ ’ß ’ø’æ’µ’°’¨’∂’•÷Ä’´ ’∂’∏÷Ä’¥’°’¨’°÷Å’¥’°’∂ ’∞’°’¥’°÷Ä:
```

```python
#Dividing into X and y
y_test = df_test.pop('price')
X_test = df_test
```

```python
# Now let's use our model to make predictions.
X_train_new = X_train_new.drop('const',axis=1)
# X_test_new ’ø’æ’µ’°’¨’∂’•÷Ä’´ ’∑÷Ä’ª’°’∂’°’Ø’´ ’Ω’ø’•’≤’Æ’∏÷Ç’¥’ù X_test-’´÷Å ÷É’∏÷É’∏’≠’°’Ø’°’∂’∂’•÷Ä’® ’∞’°’∂’•’¨’∏’æ
X_test_new = X_test[X_train_new.columns]

# ’°’æ’•’¨’°÷Å’∂’•’∂÷Ñ ’∞’°’Ω’ø’°’ø’∏÷Ç’∂ ÷É’∏÷É’∏’≠’°’Ø’°’∂’∂’•÷Ä
X_test_new = sm.add_constant(X_test_new)
```

```python
# Making predictions
y_pred = lm.predict(X_test_new)
```
